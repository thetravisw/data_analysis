By far my biggest suprise with this dataset is that first graph.  I expected a nice logarithmic curve, figuring that cardiac events in the under 40 crowd would mostly be the result of environmental stressors, while folks aged 40 and up mostly suffer heart attacks as a result of decreased cardivascular health, for which I assumed cholesterol would make a decent proxy.
